## FURTHER CLEANING

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv", dtype=str)
df
```

```python
df.drop(columns="Unnamed: 11", inplace=True)
df.info()
```

#### Date Transform

```python
INDEX_2023 = 29
INDEX_2022 = 29 + 52 * 2 + 1

df['Ngay'][:INDEX_2023] += '/2023'
df['Ngay'][INDEX_2023: INDEX_2022] += '/2022'
df['Ngay'][INDEX_2022:] += '/2021'

df['Ngay'] = pd.to_datetime(df['Ngay'], format='%d/%m/%Y')
df['Ngay']
```

#### Transform 'G4' column

```python
df['G4'] = df['G4'].apply(lambda x: x[:17] + ' ' + x[17:])
df
```

#### EXPORT

```python
df.to_csv('data.csv', index=False)
```
